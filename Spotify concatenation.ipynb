# Spotify Playlists concatenation

```python
import numpy as np
import pandas as pd
```

```python
file1=pd.read_csv("Data/hot100_Spanish_songs.csv")
file2=pd.read_csv("Data/hot1000_greatest_song.csv")
file3=pd.read_csv("Data/hiphop.csv")
file4=pd.read_csv("Data/10000songs.csv")
file5=pd.read_csv("Data/biggest_playlist.csv")
file6=pd.read_csv("Data/Spanish_songs.csv")
file7=pd.read_csv("Data/Clasic_music.csv")
file8=pd.read_csv("Data/k-pop.csv")
```

```python
df=pd.concat([file1,file2,file3,file4,file5,file6,file7,file8],axis=0).reset_index(drop=True)
```
