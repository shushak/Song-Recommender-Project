```python
#We import the credential
import config
import pandas as pd
from time import sleep
from random import randint
```

```python
#import the libraries we need 
import spotipy
import json
from spotipy.oauth2 import SpotifyClientCredentials
```

```python
#Initialize SpotiPy with user credentias

sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=config.client_id,
                                                           client_secret=config.client_secret))
```

```python
# We are going to extract the songs of the playlist:My old k-pop playlist (https://open.spotify.com/playlist/6o2lHm2gZ6TafrbiUDsFZ5)

playlist = sp.user_playlist_tracks("spotify", "6o2lHm2gZ6TafrbiUDsFZ5",market="GB")
```

```python
# we confirm the number of songs of our playlist. We have 10000 songs
print(list(playlist.keys())) # Let's look at items and total:
print("Total number of songs in the playlist: ",playlist["total"]) # 4Ksongs!! Let's check items:
len(playlist["items"])
```

```python
def get_playlist_tracks(username, playlist_id):
    results = sp.user_playlist_tracks(username,playlist_id,market="GB")
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    return tracks

```

```python
track=get_playlist_tracks("spotify","6o2lHm2gZ6TafrbiUDsFZ5")
```

```python
# function to get the songs and artist
songs=[]
artists=[]
uri=[]

for i in range(len(track)):
    uri.append(track[0]["track"]["uri"])
    songs.append(track[i]["track"]["name"])
    artists.append(track[i]["track"]["artists"][0]["name"])
```
